```python
import pandas as pd
from prefixspan import PrefixSpan
import plotly.graph_objects as go
from collections import defaultdict
```

```python
path1 = '/content/drive/MyDrive/eCommerce behavior data/sequence_event1.json'
path2 = '/content/drive/MyDrive/eCommerce behavior data/sequence_event2.json'

with open(path1, 'r') as f:
    sequence_event1 = json.load(f)
with open(path2, 'r') as f:
    sequence_event2 = json.load(f)
combined_sequence = sequence_event1 + sequence_event2
```

```python
len(combined_sequence)
```

```python
total_event = len(combined_sequence)
min_sup_percentage = 0.001
min_support_count = int(total_event * min_sup_percentage)
```

```python
ps = PrefixSpan(combined_sequence)
frequent_patterns = ps.frequent(min_support_count)
frequent_patterns.sort(key=lambda x: x[0], reverse=True)
```

```python
for freq, pattern in frequent_patterns[:15]:
    print(f"Frequency: {freq} | Pattern: {pattern}")
```

```python
business_patterns = [
    (freq, pattern) for freq, pattern in frequent_patterns
    if any('cart' in str(item) or 'purchase' in str(item) for item in pattern)
]

for freq, pattern in business_patterns[:15]:
    print(f"Frequency: {freq} | Pattern: {pattern}")
```

```python
pattern_dict = {tuple(pat): freq for freq, pat in frequent_patterns}
pattern_dict
```

```python
for freq, pat in business_patterns[:15]:
    if len(pat) > 1:
        antecedent = tuple(pat[:-1])
        consequent = pat[-1]
        if antecedent in pattern_dict:
            antecedent_freq = pattern_dict[antecedent]
            confidence = freq / antecedent_freq

            print(f"Pattern: {list(antecedent)} -> {consequent}")
            print(f"   Total Event  : {freq}")
            print(f"   Conversion Rate : {confidence:.2%} \n")
```

```python
transitions = defaultdict(int)
for freq, pat in business_patterns[:30]:
    for i in range(len(pat) - 1):
        source = pat[i]
        target = pat[i+1]
        transitions[(source, target)] += freq
nodes = list(set([src for src, tgt in transitions.keys()] + [tgt for src, tgt in transitions.keys()]))
node_indices = {node: i for i, node in enumerate(nodes)}
```

```python
sources = [node_indices[src] for src, tgt in transitions.keys()]
targets = [node_indices[tgt] for src, tgt in transitions.keys()]
values = list(transitions.values())
```

```python
fig = go.Figure(data=[go.Sankey(
    node = dict(pad = 20,thickness = 20,line = dict(color = "black", width = 0.5),label = nodes,color = "lightblue"),
    link = dict(source = sources,target = targets,value = values,color = "rgba(173, 216, 230, 0.4)" )
    )])
fig.update_layout(title_text="Customer Journey Flow (E-Commerce)", font_size=12)
fig.show()
```
